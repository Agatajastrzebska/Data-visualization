```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("city_temperature.csv")
```

```python
df2 = df1[df1['Year']<2020]
```

```python
df2['AvgTemperature'].value_counts() 
```

```python
df3 = df2[df2['AvgTemperature']> -99]
```

```python
df3['Celsjusz'] = df3['AvgTemperature'].apply(lambda x: (x-32)*(5/9))
```

```python
df3['AvgTemperature'].value_counts() 
```

```python
df4 = df3[df3['City']=="Sao Paulo"]
```

```python
df4
```

```python
df5 = df4[df4["Year"]==1996]
```

```python
df5
```

```python
df6 = df3[df3['City']=="Moscow"]
```

```python
df6
```

```python
df7 = df6[df6["Year"]==1996]
```

```python
df7
```

```python
df5['Day'] = df5['Day'].apply(lambda x: str(x)+".")
df5['Month'] = df5['Month'].apply(lambda x: str(x)+".")
df5["Date"] = df5["Day"] + df5["Month"]
df5['Date'] = pd.to_datetime(df5['Date'] + "1996", format='%d.%m.%Y')
```

```python
df5
```

```python
df7['Day'] = df7['Day'].apply(lambda x: str(x)+".")
df7['Month'] = df7['Month'].apply(lambda x: str(x)+".")
df7["Date"] = df7["Day"] + df7["Month"]
df7['Date'] = pd.to_datetime(df7['Date'] + "1996", format='%d.%m.%Y')
```

```python
dict5 = {}
dict7 = {}

for i in range(366):
    if df5["Celsjusz"].iloc[i] > df7["Celsjusz"].iloc[i]:
        dict5[abs(df5["Celsjusz"].iloc[i] - df7["Celsjusz"].iloc[i])] = [df5["Date"].iloc[i], df7["Celsjusz"].iloc[i]]
    else: 
        dict7[abs(df7["Celsjusz"].iloc[i] - df5["Celsjusz"].iloc[i])] = [df7["Date"].iloc[i], df5["Celsjusz"].iloc[i]]
```

```python
maks_temp5 = max(dict5.keys())
```

```python
maks_temp5
```

```python
maks_temp5_date = dict5[maks_temp5][0].strftime('%d.%m')
```

```python
maks_temp5_date
```

```python
max(dict7.items())
```

```python
maks_temp7 = max(dict7.keys())
```

```python
maks_temp7
```

```python
maks_temp7_date = dict7[maks_temp7][0].strftime('%d.%m')
```

```python
maks_temp7_date
```

```python
data_poczatkowa = df5['Date'].iloc[0].strftime('%d.%m.%Y')
data_koncowa = df5['Date'].iloc[-1].strftime('%d.%m.%Y')
kraj1 = df5["City"].iloc[0]
kraj2 = df7["City"].iloc[0]
```

```python
fig, axes = plt.subplots(figsize=(8,6))
plt.title("Comparison of average temperatures in {} and {}. {} - {}.".format(kraj1, kraj2, data_poczatkowa, data_koncowa))

df5.plot(x='Date', y="Celsjusz", ax=axes, label=kraj1)
df7.plot(x='Date', y="Celsjusz", ax=axes, label=kraj2)
axes.annotate("{} {:.1f}Â°C".format(maks_temp5_date, maks_temp5), xy= (dict5[maks_temp5][0], dict5[maks_temp5][1]), xytext= (dict5[maks_temp5][0], dict5[maks_temp5][1]+10), arrowprops=dict(facecolor="black", arrowstyle='-|>'))
axes.annotate("{} {:.1f}Â°C".format(maks_temp7_date, maks_temp7), xy= (dict7[maks_temp7][0], dict7[maks_temp7][1]), xytext= (dict7[maks_temp7][0], dict7[maks_temp7][1]-15), arrowprops=dict(facecolor="black", arrowstyle='-|>'))
plt.legend()
axes.spines['right'].set_color('none')
plt.ylim(-30, 40)
axes.set_ylabel('Temperature (Celsius)')

plt.savefig("zad05_Agata_Jastrzebska_281633.pdf")
plt.show()
```
