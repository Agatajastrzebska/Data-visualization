```python
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import pandas as pd 
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
df1 = pd.DataFrame(index=['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uran', 'Neptun'])
```

```python
df1["Semimajor Axis (AU)"] = [0.3870993, 0.723336, 1.000003, 1.52371, 5.2029, 9.537, 19.189, 30.0699]
```

```python
df1["Orbital Period (years)"] = [0.2408467, 0.61519726, 1.0000174, 1.8808158, 11.862615, 29.447498, 84.016846, 164.79132]
```

```python
df1
```

```python
plt.style.use('dark_background')
```

```python
fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot()
ax.plot(df1["Orbital Period (years)"],df1["Semimajor Axis (AU)"])
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Okres orbitalny - T [lata]', fontsize=15)
ax.set_ylabel('Półoś wielka - r [AU]', fontsize=15)
ax.set_title('Trzecie prawo Keplera', fontsize=15)
ax.text(15,1, r"$\frac{T^2}{r^3} = const$", fontsize=15)

planety = ['mercury.png', 'venus.png', 'earth.png', 'mars.png', 'jupiter.png', 'saturn.png', 'uran.png', 'neptun.png']
zoom = [0.07, 0.11, 0.035, 0.05, 0.08, 0.15, 0.1, 0.12]

for i in range(8):
    planeta = OffsetImage(plt.imread(planety[i]), zoom=zoom[i])
    znacznik = AnnotationBbox(planeta, (df1["Orbital Period (years)"][i], df1["Semimajor Axis (AU)"][i]), frameon=False)
    ax.add_artist(znacznik)
    
plt.savefig("zad07_Agata_Jastrzebska_281633")    
plt.show()
```
