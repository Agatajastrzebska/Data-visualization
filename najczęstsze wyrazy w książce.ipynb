```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
f = open("fatalnejaja.txt", 'r', encoding = 'utf-8')
slowa = f.read().split()
f.close()
```

```python
znaki = [':', '.', '?', ',', '/', '!', '(', ')', '"']
```

```python
for i in range(len(znaki)):
    slowa = list(map(lambda x: x.replace(znaki[i], ''), slowa))
```

```python
slowa = list(map(lambda x: x.lower(), slowa))
```

```python
slowa5 = []
```

```python
for el in slowa:
    if len(el) >= 5:
        slowa5.append(el)
```

```python
najcz_slowa = dict(Counter(slowa5).most_common(20))
```

```python
x = list(najcz_slowa.keys())
```

```python
x = x[::-1]
```

```python
y = list(najcz_slowa.values())
```

```python
y = y[::-1]
```

```python
cmap = plt.get_cmap('tab20c') 
normalize = plt.Normalize(min(y), max(y))
colors = [cmap(normalize(value)) for value in y]

plt.barh(x,y, color = colors)
plt.title('Michai Buhakow. "Fatalne jaja". Liczba sw: {}.\n20 najczciej wystpujcych sw o dugoci co najmniej 5.'.format(len(slowa)))
plt.yticks(np.arange(0, 20, 1))
plt.xticks([])

for i in range(20):
    plt.text(y[i]/2, -0.25+i*1, y[i])
plt.savefig("zad03_Agata_Jastrzebska_281633.pdf", bbox_inches='tight')
plt.show()
```
